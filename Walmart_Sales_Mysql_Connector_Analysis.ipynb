## Exploratory Data Analysis

## Univariate analysis:

### Purpose:
Examine the distribution of each feature.
### Categorical Features
Visualize the frequency of each category to understand distribution.
### Insight: 
The count plot helps to understand the most common categories within each feature.

```python
categorical_features = ['branch', 'city', 'customer_type', 'gender', 'product_line', 'payment_method', 'time_of_day', 'day_name', 'month_name']

for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=sales_df, x=feature)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

## Numerical Features:
Visualize the distribution to understand spread and central tendency.
### Insight:
The histogram shows how values are spread across the range, indicating skewness and potential outliers.

```python
numerical_features = ['unit_price', 'quantity', 'VAT', 'total', 'cogs', 'gross_margin_pct', 'gross_income', 'rating']

for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(sales_df[feature], kde=True)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

## Bivariate Analysis:

### Purpose:
Examine the relationships between pairs of features.
### Sales Performance by Branch and City
Compare sales across different branches and cities.
### Insight: 
The bar plot helps identify which branches and cities have the highest sales.

```python
sales_by_branch_city = sales_df.groupby(['branch', 'city'])['total'].sum().reset_index()
plt.figure(figsize=(8, 4))
sns.barplot(data=sales_by_branch_city, x='branch', y='total', hue='city')
plt.title("Sales Performance by Branch and City")
plt.show()
```

### Sales Performance by Product Line:
Understand which product lines contribute most to sales.
#### Insight: 
This visualization helps identify the most popular product lines driving sales.

```python
sales_by_product_line = sales_df.groupby('product_line')['total'].sum().reset_index()
plt.figure(figsize=(8, 4))
sns.barplot(data=sales_by_product_line, x='product_line', y='total')
plt.title("Sales Performance by Product Line")
plt.xticks(rotation=90)
plt.show()
```

### Customer Analysis:
Analyze sales by customer type and gender.
#### Insight: 
These bar plots help understand which customer types and genders are driving sales.

```python
sales_by_customer_type = sales_df.groupby('customer_type')['total'].sum().reset_index()
plt.figure(figsize=(14, 7))
sns.barplot(data=sales_by_customer_type, x='customer_type', y='total')
plt.title("Sales by Customer Type")
plt.show()
```

```python
sales_by_gender = sales_df.groupby('gender')['total'].sum().reset_index()
plt.figure(figsize=(14, 7))
sns.barplot(data=sales_by_gender, x='gender', y='total')
plt.title("Sales by Gender")
plt.show()
```

## Time Series Analysis
### Purpose:
Identify sales trends over time.
### Insight: 
The line plot helps identify trends, seasonality, and patterns in sales over time.

```python
sales_over_time = sales_df.groupby('date')['total'].sum().reset_index()
plt.figure(figsize=(14, 7))
sns.lineplot(data=sales_over_time, x='date', y='total')
plt.title("Sales Over Time")
plt.show()
```

## Correlation Analysis
### Purpose:
Understand relationships between numerical features.
### Insight: 
The heatmap shows the strength and direction of relationships between numerical features, 
useful for identifying potential predictors of sales.

```python
numerical_features = ['unit_price', 'quantity', 'VAT', 'total', 'cogs', 'gross_income', 'rating']
correlation_matrix = sales_df[numerical_features].corr()
plt.figure(figsize=(14, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
